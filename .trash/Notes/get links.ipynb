```python
# Print all lines in files in Downloads/Notes that contain a "[[" or "]]"
# This is a simple way to find all the links in a notebook

import os
import re

match_query = re.compile(r'\[\[.*?\]\]')

file_name_to_links = {}
output_2 = []

for filename in os.listdir('.'):
    with open(os.path.join('.', filename)) as f:
            # If the file is a markdown file
            if filename.endswith('.md'):
                lines = f.readlines()
                for line in lines:
                        # if match_query.search(line):
                        #         # Append the matching query
                        #         file_name_to_links.append(match_query.search(line).group())
                        if match_query.search(line):
                                # If the match_query does not match query 'YYYY-MM-DD' in regex
                                if not re.search(r'\[\[\d{4}-\d{2}-\d{2}\]\]', match_query.search(line).group()):
                                        if filename not in file_name_to_links:
                                                file_name_to_links[filename] = []
                                        file_name_to_links[filename].append(match_query.search(line).group())
# Sort file_name_to_links by key
file_name_to_links = {k: v for k, v in sorted(file_name_to_links.items(), key=lambda item: item[0])}
                       
```

```python
import json
 # Write the output to a file
with open('file_name_to_links.json', 'w') as f:
	json.dump(file_name_to_links, f)
```

```python
# File names not in output_1
files_not_in_output_1 = [filename for filename in os.listdir('.') if filename not in file_name_to_links]
files_not_in_output_1.sort()
# Move the files to ./NoLinks/
for filename in files_not_in_output_1:
	os.rename(os.path.join('.', filename), os.path.join('.', 'NoLinks', filename))
```
